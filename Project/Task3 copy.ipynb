#### –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ
–î–æ–±–∞–≤—å—Ç–µ –∫ —Å—É—â–µ—Å—Ç—É—é—â–∏–º –∫–ª–∞—Å—Å–∞–º –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä–∞ –∫–ª–∞—Å—Å–æ–≤ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –≥–µ—Ç—Ç–æ—Ä–æ–≤ –∏ —Å–µ—Ç—Ç–æ—Ä–æ–≤ –∏ —Ä–µ–∞–ª–∏–∑—É–π—Ç–µ –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏–µ –æ–±—ä–µ–∫—Ç–æ–≤ –º–µ–∂–¥—É —Å–æ–±–æ–π

```
using System;
using System.Collections.Generic;

// 1. –ö–ª–∞—Å—Å –¢–æ–≤–∞—Ä (Product) —Å —É–ª—É—á—à–µ–Ω–Ω–æ–π –ª–æ–≥–∏–∫–æ–π –∏ –≥–µ—Ç—Ç–µ—Ä–∞–º–∏/—Å–µ—Ç—Ç–µ—Ä–∞–º–∏
public class Product
{
    private string name;
    private decimal price;
    private int quantity;
    protected string category;
    internal string description;

    public Product(string name, decimal price, int quantity, string category, string description)
    {
        this.name = name;
        this.price = price;
        this.quantity = quantity;
        this.category = category;
        this.description = description;
    }

    // –ì–µ—Ç—Ç–µ—Ä—ã –∏ —Å–µ—Ç—Ç–µ—Ä—ã
    public string Name
    {
        get { return name; }
        set { name = value; }
    }

    public decimal Price
    {
        get { return price; }
        set { price = value < 0 ? 0 : value; } // –ó–∞—â–∏—Ç–∞ –æ—Ç –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π —Ü–µ–Ω—ã
    }

    public int Quantity
    {
        get { return quantity; }
        set { quantity = value < 0 ? 0 : value; } // –ó–∞—â–∏—Ç–∞ –æ—Ç –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞
    }

    public string Category
    {
        get { return category; }
        set { category = value; }
    }

    public string Description
    {
        get { return description; }
        set { description = value; }
    }

    public decimal GetTotalPrice()
    {
        return price * quantity;
    }

    public void UpdateQuantity(int amount)
    {
        quantity += amount;
        if (quantity < 0) quantity = 0; // –ó–∞—â–∏—Ç–∞ –æ—Ç –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞
        Console.WriteLine($"üîÑ –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ–≤–∞—Ä–∞ '{name}' –æ–±–Ω–æ–≤–ª–µ–Ω–æ: {quantity} —à—Ç.");
    }

    public string GetProductInfo()
    {
        return $" –¢–æ–≤–∞—Ä: {name} |  –¶–µ–Ω–∞: {price:C} |  –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ: {quantity} |  –ö–∞—Ç–µ–≥–æ—Ä–∏—è: {category} |  –û–±—â–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å: {GetTotalPrice():C}";
    }
}

// 2. –ö–ª–∞—Å—Å –ü—Ä–æ–¥–∞–≤–µ—Ü (Seller) —Å —É–ª—É—á—à–µ–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω–æ—Å—Ç—å—é
public class Seller
{
    private string name;
    private string employeeId;
    protected decimal salary;
    internal string contactInfo;
    private List<Product> products = new List<Product>();

    public Seller(string name, string employeeId, decimal salary, string contactInfo)
    {
        this.name = name;
        this.employeeId = employeeId;
        this.salary = salary;
        this.contactInfo = contactInfo;
    }

    // –ì–µ—Ç—Ç–µ—Ä—ã –∏ —Å–µ—Ç—Ç–µ—Ä—ã
    public string Name
    {
        get { return name; }
        set { name = value; }
    }

    public string EmployeeId
    {
        get { return employeeId; }
        set { employeeId = value; }
    }

    public decimal Salary
    {
        get { return salary; }
        set { salary = value < 0 ? 0 : value; }
    }

    public void AddProduct(Product product)
    {
        products.Add(product);
        Console.WriteLine($" –¢–æ–≤–∞—Ä '{product.Name}' –¥–æ–±–∞–≤–ª–µ–Ω –ø—Ä–æ–¥–∞–≤—Ü—É {name}");
    }

    public void SellProduct(Product product, int quantity)
    {
        if (product.Quantity >= quantity)
        {
            product.UpdateQuantity(-quantity);
            Console.WriteLine($" –ü—Ä–æ–¥–∞–Ω–æ {quantity} —à—Ç. —Ç–æ–≤–∞—Ä–∞ '{product.Name}' –ø—Ä–æ–¥–∞–≤—Ü–æ–º {name}");
        }
        else
        {
            Console.WriteLine($" –ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Ç–æ–≤–∞—Ä–∞ '{product.Name}' –Ω–∞ —Å–∫–ª–∞–¥–µ. –î–æ—Å—Ç—É–ø–Ω–æ: {product.Quantity}, –ó–∞–ø—Ä–æ—à–µ–Ω–æ: {quantity}");
        }
    }

    public string GetSellerInfo()
    {
        return $" –ü—Ä–æ–¥–∞–≤–µ—Ü: {name} |  ID: {employeeId} |  –ó–∞—Ä–ø–ª–∞—Ç–∞: {salary:C} |  –ö–æ–Ω—Ç–∞–∫—Ç—ã: {contactInfo} |  –¢–æ–≤–∞—Ä–æ–≤ –≤ —É–ø—Ä–∞–≤–ª–µ–Ω–∏–∏: {products.Count}";
    }

    public void ListSellerProducts()
    {
        Console.WriteLine($"\n –¢–æ–≤–∞—Ä—ã –ø—Ä–æ–¥–∞–≤—Ü–∞ {name}:");
        if (products.Count == 0)
        {
            Console.WriteLine("   –¢–æ–≤–∞—Ä–æ–≤ –Ω–µ—Ç");
            return;
        }
        
        foreach (var product in products)
        {
            Console.WriteLine($"   - {product.GetProductInfo()}");
        }
    }

    public List<Product> GetProducts()
    {
        return products;
    }
}

// 3. –ö–ª–∞—Å—Å –ú–∞–≥–∞–∑–∏–Ω (Store) —Å —Ä–∞—Å—à–∏—Ä–µ–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω–æ—Å—Ç—å—é
public class Store
{
    private string storeName;
    private string location;
    public string storeHours;
    private List<Seller> sellers = new List<Seller>();
    private List<Product> storeProducts = new List<Product>();

    public Store(string storeName, string location, string storeHours)
    {
        this.storeName = storeName;
        this.location = location;
        this.storeHours = storeHours;
    }

    // –ì–µ—Ç—Ç–µ—Ä—ã –∏ —Å–µ—Ç—Ç–µ—Ä—ã
    public string StoreName
    {
        get { return storeName; }
        set { storeName = value; }
    }

    public string Location
    {
        get { return location; }
        set { location = value; }
    }

    public void AddSeller(Seller seller)
    {
        sellers.Add(seller);
        Console.WriteLine($" –ü—Ä–æ–¥–∞–≤–µ—Ü {seller.Name} –¥–æ–±–∞–≤–ª–µ–Ω –≤ –º–∞–≥–∞–∑–∏–Ω '{storeName}'");
    }

    public void AddProductToStore(Product product)
    {
        storeProducts.Add(product);
        Console.WriteLine($" –¢–æ–≤–∞—Ä '{product.Name}' –¥–æ–±–∞–≤–ª–µ–Ω –≤ –º–∞–≥–∞–∑–∏–Ω");
    }

    public void ListProducts()
    {
        Console.WriteLine($"\n –í–°–ï –¢–û–í–ê–†–´ –í –ú–ê–ì–ê–ó–ò–ù–ï '{storeName}':");
        if (storeProducts.Count == 0 && sellers.Count == 0)
        {
            Console.WriteLine("   –¢–æ–≤–∞—Ä–æ–≤ –Ω–µ—Ç");
            return;
        }

        // –¢–æ–≤–∞—Ä—ã –º–∞–≥–∞–∑–∏–Ω–∞
        if (storeProducts.Count > 0)
        {
            Console.WriteLine("\n –¢–æ–≤–∞—Ä—ã –º–∞–≥–∞–∑–∏–Ω–∞:");
            foreach (var product in storeProducts)
            {
                Console.WriteLine($"   - {product.GetProductInfo()}");
            }
        }

        // –¢–æ–≤–∞—Ä—ã –ø—Ä–æ–¥–∞–≤—Ü–æ–≤
        if (sellers.Count > 0)
        {
            Console.WriteLine("\n –¢–æ–≤–∞—Ä—ã –ø—Ä–æ–¥–∞–≤—Ü–æ–≤:");
            foreach (var seller in sellers)
            {
                Console.WriteLine($"   –ü—Ä–æ–¥–∞–≤–µ—Ü: {seller.Name}");
                foreach (var product in seller.GetProducts())
                {
                    Console.WriteLine($"     - {product.GetProductInfo()}");
                }
            }
        }
    }

    public void ListSellers()
    {
        Console.WriteLine($"\n –ü–†–û–î–ê–í–¶–´ –ú–ê–ì–ê–ó–ò–ù–ê '{storeName}':");
        if (sellers.Count == 0)
        {
            Console.WriteLine("   –ü—Ä–æ–¥–∞–≤—Ü–æ–≤ –Ω–µ—Ç");
            return;
        }
        
        foreach (var seller in sellers)
        {
            Console.WriteLine($"   - {seller.GetSellerInfo()}");
        }
    }

    public string GetStoreInfo()
    {
        return $"–ú–∞–≥–∞–∑–∏–Ω: {storeName} |  –ê–¥—Ä–µ—Å: {location} |  –ß–∞—Å—ã —Ä–∞–±–æ—Ç—ã: {storeHours} |  –ü—Ä–æ–¥–∞–≤—Ü–æ–≤: {sellers.Count} |  –¢–æ–≤–∞—Ä–æ–≤: {storeProducts.Count}";
    }
}

// –î–ï–ú–û–ù–°–¢–†–ê–¶–ò–Ø –†–ê–ë–û–¢–´ –ö–õ–ê–°–°–û–í
Console.WriteLine("===  –î–ï–ú–û–ù–°–¢–†–ê–¶–ò–Ø –†–ê–ë–û–¢–´ –°–ò–°–¢–ï–ú–´ –ú–ê–ì–ê–ó–ò–ù–ê ===\n");

// –°–æ–∑–¥–∞–Ω–∏–µ —ç–∫–∑–µ–º–ø–ª—è—Ä–æ–≤ –∫–ª–∞—Å—Å–æ–≤
Product product1 = new Product("–ê–ø–µ–ª—å—Å–∏–Ω—ã", 1.20m, 100, "–§—Ä—É–∫—Ç—ã", "–°–æ—á–Ω—ã–µ –∞–ø–µ–ª—å—Å–∏–Ω—ã");
Product product2 = new Product("–Ø–±–ª–æ–∫–∏", 0.80m, 150, "–§—Ä—É–∫—Ç—ã", "–°–≤–µ–∂–∏–µ —è–±–ª–æ–∫–∏");
Product product3 = new Product("–ú–æ–ª–æ–∫–æ", 1.50m, 50, "–ú–æ–ª–æ—á–Ω—ã–µ –ø—Ä–æ–¥—É–∫—Ç—ã", "–°–≤–µ–∂–µ–µ –º–æ–ª–æ–∫–æ");

Seller seller1 = new Seller("–ù–∏–∫–∏—Ç–∞", "–ö1237", 3000.00m, "tresh78803@gmail.com");
Seller seller2 = new Seller("–ê–Ω–Ω–∞", "–ö1238", 3200.00m, "anna@store.com");

Store store = new Store("–ü—Ä–æ–¥—É–∫—Ç—ã", "–¶–µ–Ω—Ç—Ä –≥–æ—Ä–æ–¥–∞", "–ö—Ä—É–≥–ª–æ—Å—É—Ç–æ—á–Ω–æ");

// –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –≤ –º–∞–≥–∞–∑–∏–Ω
store.AddSeller(seller1);
store.AddSeller(seller2);

// –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Ç–æ–≤–∞—Ä–æ–≤ –ø—Ä–æ–¥–∞–≤—Ü–∞–º
seller1.AddProduct(product1);
seller1.AddProduct(product2);
seller2.AddProduct(product3);

// –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Ç–æ–≤–∞—Ä–æ–≤ –Ω–µ–ø–æ—Å—Ä–µ–¥—Å—Ç–≤–µ–Ω–Ω–æ –≤ –º–∞–≥–∞–∑–∏–Ω
store.AddProductToStore(new Product("–•–ª–µ–±", 0.90m, 200, "–í—ã–ø–µ—á–∫–∞", "–°–≤–µ–∂–∏–π —Ö–ª–µ–±"));

// –í—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏
Console.WriteLine("\n" + new string('=', 50));
Console.WriteLine(store.GetStoreInfo());
Console.WriteLine(new string('=', 50));

Console.WriteLine("\n" + new string('=', 50));
store.ListSellers();
Console.WriteLine(new string('=', 50));

Console.WriteLine("\n" + new string('=', 50));
store.ListProducts();
Console.WriteLine(new string('=', 50));

// –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è –ø—Ä–æ–¥–∞–∂–∏ —Ç–æ–≤–∞—Ä–æ–≤
Console.WriteLine("\n" + new string('=', 50));
Console.WriteLine(" –î–ï–ú–û–ù–°–¢–†–ê–¶–ò–Ø –ü–†–û–î–ê–ñ:");
Console.WriteLine(new string('=', 50));
seller1.SellProduct(product1, 10);
seller2.SellProduct(product3, 5);
seller1.SellProduct(product2, 200); // –ü–æ–ø—ã—Ç–∫–∞ –ø—Ä–æ–¥–∞—Ç—å –±–æ–ª—å—à–µ —á–µ–º –µ—Å—Ç—å

// –í—ã–≤–æ–¥ –∏—Ç–æ–≥–æ–≤–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –ø–æ—Å–ª–µ –ø—Ä–æ–¥–∞–∂
Console.WriteLine("\n" + new string('=', 50));
Console.WriteLine("–ò–¢–û–ì–û–í–ê–Ø –ò–ù–§–û–†–ú–ê–¶–ò–Ø –ü–û–°–õ–ï –ü–†–û–î–ê–ñ:");
Console.WriteLine(new string('=', 50));
store.ListProducts();

// –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è —Ä–∞–±–æ—Ç—ã –≥–µ—Ç—Ç–µ—Ä–æ–≤ –∏ —Å–µ—Ç—Ç–µ—Ä–æ–≤
Console.WriteLine("\n" + new string('=', 50));
Console.WriteLine("–î–ï–ú–û–ù–°–¢–†–ê–¶–ò–Ø –ì–ï–¢–¢–ï–†–û–í –ò –°–ï–¢–¢–ï–†–û–í:");
Console.WriteLine(new string('=', 50));
product1.Price = 1.50m; // –ò–∑–º–µ–Ω–µ–Ω–∏–µ —Ü–µ–Ω—ã —á–µ—Ä–µ–∑ —Å–µ—Ç—Ç–µ—Ä
product1.Quantity = 95; // –ò–∑–º–µ–Ω–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —á–µ—Ä–µ–∑ —Å–µ—Ç—Ç–µ—Ä
Console.WriteLine($"–û–±–Ω–æ–≤–ª–µ–Ω–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ç–æ–≤–∞—Ä–µ: {product1.GetProductInfo()}");
```
